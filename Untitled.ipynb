```python
import os
from pathlib import Path
import pandas as pd

for file in Path('./data/gpt2/').glob('**/*.txt'):
    df = pd.read_csv(file, 
                     sep = '\n', 
                     header=None, 
                     error_bad_lines=False, 
                     names=['text'])

df.head()
df.shape
```

```python
from nltk import word_tokenize

def preprocess_text(df):
    # Define characters to remove 
    delete_table = dict.fromkeys(map(ord, '<|>'), '')
    df.text = df.text.apply(lambda sent: sent.translate(delete_table).replace('startoftext', '').replace('endoftext', ''))
    df['text'].dropna(inplace=True) # Remove empty rows
    
    # Drop singe word sentences
    df['length'] = df.text.apply(lambda sent: len(word_tokenize(sent)))
    df = df[df['length'] > 1]
    df.drop('length', axis=1, inplace=True)
    
    return df

df = preprocess_text(df)
df.shape
```

```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
import string

def clean_text(sent):
    # Convert to lowercase
    sent = sent.lower()
    
    # Remove punctuation
    translation = dict.fromkeys(map(ord, string.punctuation), '')
    sent = sent.translate(translation)
    
    return sent
```

```python
df = df.reset_index()
df['cleaned'] = df.text.apply(lambda sent: clean_text(sent))
df
```

```python
def remove_duplicates(df, /, percentage=0.9):
    # Remove exact duplicates.
    df['cleaned'] = df['text'].apply(lambda x: clean_text(x))
    df.drop_duplicates(['cleaned'], keep='first', inplace=True)
    print('After dropping duplicates:', len(df))

    # Remove very similar texts.
    indices = []

    for row in df.itertuples():
        for row2 in df.loc[row[0] + 1:].itertuples():
            if similar(row[3], row2[3]) >= percentage:
                indices.append((row[0], row2[0]))

    similars = [first for first, second in list(set(tuple(sorted(sub)) for sub in indices))]
    df.drop(similars, inplace=True)

    # Dropped a total of 30.
    return df
```

```python
file = './data/gibert/hate_train.csv'
og_df = pd.read_csv(file, sep = '\n', header=None, error_bad_lines = False, names=['text'])
og_df
```

```python
def remove_replicates(new_df, og_df, /, percentage=0.9):
    df = new_df.append(og_df)
    #df = remove_duplicates(df)
    df['cleaned'] = df['text'].apply(lambda x: clean_text(x))
    df.drop_duplicates(['cleaned'], keep='first', inplace=True)
    df.dropna('index', inplace=True) # Remove empty rows
    df.drop(['cleaned'], axis=1, inplace=True)

    print('After dropping duplicates:', len(df))
    return df
```

```python
final = remove_replicates(df, og_df)
print(final.shape)
final.to_csv('./data/gpt2/cleaned_hate.csv', index=False)
```
